```python
import pandas as pd 

df = pd.read_csv("datasets/pokemon_data.txt", delimiter='\t')
```

```python
df.head(5)
```

CREATE TOTAL 

```python
# Create total by adding the selected fields 
df["Total"] = df["HP"] + df["Attack"] + df["Defense"] + df["Sp. Atk"] + df["Sp. Def"] + df["Speed"] 

df.head(5)
```

```python
# Drop (delete) the total column
df = df.drop(columns=["Total"])

df.head(5)
```

```python
# Add the total by adding a range of columns numbers. Notes: (1) the upper bound will not be included (2) axis=1: horizontally. Warning! remember to put a comma between the row and column  
df["Total"] = df.iloc[ :, 4:10].sum(axis=1)  # Adding columns 4 to 9, but not including column 10

df.head(5)
```

```python
# Change the location of total to column 4, that is right after "Type 2". Note: cols[-1] is the last column 
cols = list(df.columns.values)
df = df[cols[0:4] +[cols[-1]] + cols[4:12]]

df.head(5)
```

SAVING DATA IN DF TO FILES 

```python
# Saving data without the index to csv file 
df.to_csv("datasets/modified_pokemon.csv", index=False)
```

```python
# Saving data without the index to excel xlsx file 
df.to_excel("datasets/modified_pokemon.xlsx", index=False)
```

```python
# Saving data without the index to text delimiter file 
df.to_csv("datasets/modified_pokemon.txt", index=False, sep='\t')
```

```python
FILTERING DATA 
```

```python
# Select only data with Type 1 = "Grass"
df.loc[df["Type 1"] == "Grass"]
```

```python
# Select only data with Type 1 = "Grass" and Type 2 = "Poison". Note: only one & for logical and 
df.loc[(df["Type 1"] == "Grass") & (df["Type 2"] == "Poison")]
```

```python
# Select only data with either Type 1 = "Grass" OR Type 2 = "Poison". Note: only one | for logical OR 
df.loc[(df["Type 1"] == "Grass") | (df["Type 2"] == "Poison")]
```

```python
new_df = df.loc[(df["Type 1"] == "Grass") & (df["Type 2"] == "Poison")]

# create a new ascending index and remove the old index 
new_df = new_df.reset_index(drop=True)
new_df.head(5)
```

```python
# Select only data having "Name" containing the word "Mega"
df.loc[df["Name"].str.contains("Mega")]
```

```python
# Select only data having "Name" excludes the word "Mega". Note: use "~" for exclusion 
df.loc[~df["Name"].str.contains("Mega")]
```

```python
# Use regular expression for filtering data 
import re   # regular expression library 

# Select only data having Type 1 contains either the word "Fire" or "Grass"
#  Notes: use "re.I" to ignore case sensitive data 
df.loc[df["Type 1"].str.contains("fire|grass", flags=re.I, regex=True)]
```

```python
# Select only data having Name contains "pi" with (zero or more) letters followed
#  Notes: use "re.I" to ignore case sensitive data 
df.loc[df["Name"].str.contains("pi[a+z]*", flags=re.I, regex=True)]
```

```python
# Select only data having Name begins with "pi" with (zero or more) letters followed
#  Notes: use "re.I" to ignore case sensitive data 
df.loc[df["Name"].str.contains("^pi[a+z]*", flags=re.I, regex=True)]
```

CONDIITIONAL CHANGES 

```python
# Read "modified_pokemon.csv" into data frame 
import pandas as pd 

df = pd.read_csv("datasets/modified_pokemon.csv")

# Conditional changes by setting "Legendary" to True if "Type 1" is Grass
df.loc[df["Type 1"] == "Grass", "Legendary"] = True

df.head(5)
```

```python
# Read "modified_pokemon.csv" into data frame 
import pandas as pd 

df = pd.read_csv("datasets/modified_pokemon.csv")

# Set "Generation" & "Legendary" to "TEST VALUE"
df.loc[df["Total"] > 500, ["Generation", "Legendary"]] = ["Test 1", "Test 2"]

df.head(5)
```

AGGREGATE STATISTICS (groupby)

```python
# Read "modified_pokemon.csv" into data frame 
import pandas as pd 

df = pd.read_csv("datasets/modified_pokemon.csv")

df.groupby(["Type 1"]).mean().sort_values("Attack", ascending=False)
```

```python
df.groupby(["Type 1"]).sum()
```

```python
df.groupby(["Type 1"]).count()
```

```python
df["count"] = 1
df.groupby(["Type 1", "Type 2"]).count()["count"]
```

WORKING WITH LARGE AMOUNT OF DATA 

```python
for df in pd.read_csv("datasets/modified_pokemon.csv", chunksize=5):
    print("CHUNKED DF\n")
    print(df)
```

```python
new_df = pd.DataFrame(columns=df.columns)

for df in pd.read_csv("datasets/modified_pokemon.csv", chunksize=5):
    results = df.groupby(["Type 1"]).count()
    new_df = pd.concat([new_df, results])
    
    print(new_df)

```
